```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,5)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.drop_duplicates(inplace=True)
df.head()
```

```python
df['city'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Cities on Swiggy")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(y=df["food_type"], order=df["food_type"].value_counts().head(10).index)
plt.title("Top Food Categories")
plt.show()
```

```python
sns.histplot(df["avg_ratings"], kde=True, bins=25)
plt.title("Ratings Distribution")
plt.show()
```

```python
sns.histplot(df["delivery_time"], kde=True, bins=25)
plt.title("Delivery Time Distribution")
plt.show()
```
